```
import pandas as pd
import csv
import re

with open('covid19_vaccine_tweets.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)

columns = data.pop(0)
tweetIdList = []
print(len(data))
removelist = ".!@/:# "
for x in data:
  id = x[0]
  tweet = x[1]
  tweet = remove_emojis(tweet)
  tweet = re.sub(r'[^\w'+removelist+']', '', tweet)
  combined = id + ' tweet: ' + tweet
  tweetIdList.append(combined)

print(tweetIdList)
# need to split lists into groups of 12
tweetSplit = list(chunks(tweetIdList, 10))
```

```
def chunks(l, n):
    n = max(1, n)
    return (l[i:i+n] for i in range(0, len(l), n))
```

```

tweetColumns = ['tweet1', 'tweet2', 'tweet3', 'tweet4', 'tweet5', 'tweet6', 'tweet7', 'tweet8', 'tweet9', 'tweet10']

# replace [sacramenta, california] with split lists
pdData = pd.DataFrame(tweetSplit, columns=tweetColumns)
pdData.to_csv('combinedTweets-update.csv', index=False)
```

```
import re
def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)
```
